```python
count_neg = 0

files = glob.glob('/kaggle/input/rsna-breast-cancer-1024-pngs/output/*.png')
DESTINATION_PATH = '/kaggle/working/negative'  # The preferred path for saving the processed image
os.mkdir(DESTINATION_PATH)
#directory_images = '/kaggle/input/rsna-breast-cancer-detection-png256nearest/output'
for f in files:
    fn_name = f.split('_')[-1][:-4]
    img = Image.open(f)
    if count_neg!=1018:
            try:
                if ((df['image_id']==int(fn_name))& (df['cancer']==0)& (df['difficult_negative_case'] == False)).any():
                    base_filename = os.path.basename(f)
                    title, ext = os.path.splitext(base_filename)
                    final_filepath = os.path.join(DESTINATION_PATH, fn_name + ext)
                    img.save(final_filepath)
                    #image_path = '/kaggle/working/negative'
                    #os.mkdir(image_path)
                    #image = img.save(f"{image_path}/{fn_name}.png")
                    count_neg=count_neg+1
            except:
                pass
    else:
        break
```
